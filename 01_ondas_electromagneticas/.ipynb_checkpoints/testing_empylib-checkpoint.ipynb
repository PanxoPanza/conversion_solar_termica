```python
import matplotlib.pyplot as plt
import numpy as np
import empylib.nklib as nk

lam = np.logspace(np.log10(0.2),np.log10(55), 1000)
N = nk.ZnS(lam)

fig, ax = plt.subplots()
ax.plot(lam, N.real, "-", color="#aa0044")
ax.plot(lam, N.imag, "-", color="#6886b3")
ax.set_xlabel(r"Wavelength ($\rm μm$)")
ax.set_xscale('log')
plt.show()
```
