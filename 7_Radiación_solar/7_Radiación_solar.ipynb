## Características del sol

El sol es una esfera de materia gaseosa de aproximadamente $1.39\times10^9$m de diámetro.


<img src="./images/sun_structure.png" width="400px" align= center>


La radiación solar puede ser estimada como un cuerpo negro a temperatura $T_\mathrm{sun} = 5777 K$. Sin embargo, la radiación solar es el resultado de la emisión de muchas capas con diferente composición, y cuya temperatura puede alcanzar más de 100,000 K.

### Constante solar

La figura muestra un esquema de la relación geométrica del sol y la tierra. La distancia entre el sol y la tierra es de $1.495\times 10^{11}\pm 1.7$%. La variación de 1.7% está asociada a la excentricidad de la órbita de la tierra.

<img src="./images/solar_constant.png" width="400px" align= center>

La **constante solar $G_{sc} = 1367$ W/m$^2$** es la irradiancia total sobre una superficie perpendicular a los rayos solares y a la distancia media entre la tierra y el sol. En otras palabras, corresponde a la energía total proveniente del sol que incide fuera de la atmosfera de la tierra.
